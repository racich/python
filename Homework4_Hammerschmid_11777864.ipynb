#### Momentum and trading volume
In this exercise, you are asked to obtain some stock market data and to compute and plot returns of certain investment strategies.

1. Please download data (over the whole available time period) for the stock symbols CSCO, GE, F, MSFT, AIG, JPM, AMGN, BRK-A, IBM, and AA for the period 1995 to 2020. Plot the cumulative returns of all of these stocks in a single figure.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as web

pd.set_option('display.max_rows', 20000)

#Downloading the data
CSCO = web.DataReader("CSCO", 'yahoo',start='1995', end='2020')
GE = web.DataReader("GE", 'yahoo', start='1995', end='2020')
F = web.DataReader("F", 'yahoo', start='1995', end='2020')
MSFT = web.DataReader("MSFT", 'yahoo', start='1995', end='2020')
AIG = web.DataReader("AIG", 'yahoo', start='1995', end='2020')
JPM = web.DataReader("JPM", 'yahoo', start='1995', end='2020')
AMGN = web.DataReader("AMGN", 'yahoo', start='1995', end='2020')
BRKA = web.DataReader("BRK-A", 'yahoo', start='1995', end='2020')
IBM = web.DataReader("IBM", 'yahoo', start='1995', end='2020')
AA = web.DataReader("AA", 'yahoo', start='1995', end='2020')



```

```python
#Rename "Adj Close" column to the name of the stock
CSCO = CSCO.rename(columns={'Adj Close': 'CSCO'})
GE = GE.rename(columns={'Adj Close': 'GE'})
F = F.rename(columns={'Adj Close': 'F'})
MSFT = MSFT.rename(columns={'Adj Close': 'MSFT'})
AIG = AIG.rename(columns={'Adj Close': 'AIG'})
JPM = JPM.rename(columns={'Adj Close': 'JPM'})
AMGN = AMGN.rename(columns={'Adj Close': 'AMGN'})
BRKA = BRKA.rename(columns={'Adj Close': 'BRKA'})
IBM = IBM.rename(columns={'Adj Close': 'IBM'})
AA = AA.rename(columns={'Adj Close': 'AA'})

#Finding cummulative returns
CSCO_cum = (CSCO['CSCO'].pct_change()+1).cumprod()
GE_cum = (GE['GE'].pct_change()+1).cumprod()
F_cum = (F['F'].pct_change()+1).cumprod()
MSFT_cum = (MSFT['MSFT'].pct_change()+1).cumprod()
AIG_cum = (AIG['AIG'].pct_change()+1).cumprod()
JPM_cum = (JPM['JPM'].pct_change()+1).cumprod()
AMGN_cum = (AMGN['AMGN'].pct_change()+1).cumprod()
BRKA_cum = (BRKA['BRKA'].pct_change()+1).cumprod()
IBM_cum = (IBM['IBM'].pct_change()+1).cumprod()
AA_cum = (AA['AA'].pct_change()+1).cumprod()

#Combine all the data into one dataframe
allstocks = pd.concat([CSCO_cum, GE_cum, F_cum,MSFT_cum, AIG_cum, JPM_cum, AMGN_cum, BRKA_cum, IBM_cum, AA_cum], axis = 1)

#Plot all stocks' cummulative returns in one graph
allstocks.plot(figsize= (18, 6))
plt.title('Cummulative Returns of All Stocks', fontsize='16')
plt.xlabel('Date', fontsize='14')
plt.ylabel('Growth of $1 Investment', fontsize='14');
```

2. The goal now is to first consider returns of simple momentum/reversal investment strategies. A momentaum (reversal) strategy invests in recent winners (losers) and takes a short position in recent losers (winners). Assume that whenever you change your portfolio (i.e. at the opening price on every day, week, or month), you invest the same amount in each position, and compute the returns of strategies that

* every day take a long position in the previous day's 5 worst performing stocks and a short position in the previous day's 5 best performing stocks.

* every week take a long position in the previous week's 5 worst performing stocks and a short position in the previous week's 5 best performing stocks.

* every month __t__ take a long position in the 5 stocks performing best during the months __t-6__ to __t-2__ and a short position in the 5 stocks performing worst during that period.

Plot the cumulative returns of each of these strategies and provide descriptive statistics of the one-period returns.

```python
stocks = pd.concat([CSCO['CSCO'], GE['GE'], F['F'],MSFT['MSFT'], AIG['AIG'], JPM['JPM'], AMGN['AMGN'], BRKA['BRKA'], IBM['IBM'], AA['AA']], axis = 1)
stocks.dropna(inplace=True)
returns = (stocks / stocks.shift(1)).iloc[1:]
cols = returns.columns


for i in range(1, len(returns)):
    pre_day = returns.iloc[i-1,:][cols]
    best_stocks = pre_day.sort_values(ascending=False)[:5].index
    worst_stocks = pre_day.sort_values(ascending=False)[5:].index
    daily_update_long_returns = returns.iloc[i,:][worst_stocks]
    daily_update_short_returns = 1 - (returns.iloc[i,:][best_stocks]-1)
    total_daily_update_returns = pd.concat([daily_update_long_returns, daily_update_short_returns])
    returns.loc[returns.index[i], 'daily_update'] = total_daily_update_returns.mean()

returns.head()
```

```python
print(returns['daily_update'].describe())
returns['daily_update'].cumprod().plot()
```

```python
wkl_ret=(stocks / stocks.shift(1)).iloc[1:].resample("W").prod()

for i in range(1, len(wkl_ret)):
    pre_w = wkl_ret.iloc[i-1,:][cols]
    best_stocks = pre_w.sort_values(ascending=False)[:5].index
    worst_stocks = pre_w.sort_values(ascending=False)[5:].index
    w_update_long_returns = wkl_ret.iloc[i,:][worst_stocks]
    w_update_short_returns = 1 - (wkl_ret.iloc[i,:][best_stocks]-1)
    total_w_update_returns = pd.concat([w_update_long_returns, w_update_short_returns])
    wkl_ret.loc[wkl_ret.index[i], 'w_update'] = total_w_update_returns.mean()
    
print(wkl_ret.head())
```

```python
wkl_ret['w_update'].cumprod().plot()
print(wkl_ret['w_update'].describe())
```

```python
mtl_ret=(stocks / stocks.shift(1)).iloc[1:].resample("M").prod()



for i in range(5, len(mtl_ret)):
    pre_m = mtl_ret.iloc[i-6:i-2,:][cols].prod()
    best_stocks = pre_m.sort_values(ascending=False)[:5].index
    worst_stocks = pre_m.sort_values(ascending=False)[5:].index
    m_update_long_returns = mtl_ret.iloc[i,:][best_stocks]
    m_update_short_returns = 1 - (mtl_ret.iloc[i,:][worst_stocks]-1)
    total_m_update_returns = pd.concat([m_update_long_returns, m_update_short_returns])
    mtl_ret.loc[mtl_ret.index[i], 'm_update'] = total_m_update_returns.mean()
```

```python
mtl_ret['m_update'].cumprod().plot()
print(mtl_ret['m_update'].describe())
```

3. We next consider a refinement of the above strategies that takes the trading volume in each stock, relative to the stocks' own historical volume, into account. To prepare for this step, compute for each stock

* every day the trading volume (in shares) on that day relative to the trading volume on the previous 10 trading days
* every week the trading volume (in shares) in that week relative to the trading volume during the previous 4 weeks
* every month the trading volume (in shares) during the months __t-6__ to __t-2__ relative to the trading volume in the previous 10 months

Next, adjust the investment strategies from 2. such that you always take long and short positions, respectively, in the 2 among the 5 stocks that had the highest recent trading volume (relative to their past volumes, computed as above).

Plot the cumulative returns of each of these strategies and provide descriptive statistics of the one-period returns.

```python
CSCO = CSCO.rename(columns={'Volume': 'CSCOv'})
GE = GE.rename(columns={'Volume': 'GEv'})
F = F.rename(columns={'Volume': 'Fv'})
MSFT = MSFT.rename(columns={'Volume': 'MSFTv'})
AIG = AIG.rename(columns={'Volume': 'AIGv'})
JPM = JPM.rename(columns={'Volume': 'JPMv'})
AMGN = AMGN.rename(columns={'Volume': 'AMGNv'})
BRKA = BRKA.rename(columns={'Volume': 'BRKAv'})
IBM = IBM.rename(columns={'Volume': 'IBMv'})
AA = AA.rename(columns={'Volume': 'AAv'})

volume = pd.concat([CSCO['CSCOv'], GE['GEv'], F['Fv'],MSFT['MSFTv'], AIG['AIGv'], JPM['JPMv'], AMGN['AMGNv'], BRKA['BRKAv'], IBM['IBMv'], AA['AAv']], axis = 1)
volume.columns=['CSCO', 'GE', 'F','MSFT', 'AIG', 'JPM', 'AMGN', 'BRKA', 'IBM', 'AA']
volume.head()
```

```python
#daily volume
for i in range(10, len(returns)):
    pre_day = returns.iloc[i-1,:][cols]
    best_stocks = pre_day.sort_values(ascending=False)[:5].index
    worst_stocks = pre_day.sort_values(ascending=False)[5:].index
    vol_change_best=((volume[i-11:i-1][best_stocks].mean())/volume.iloc[i][best_stocks])-1
    vol_change_worst=((volume[i-11:i-1][worst_stocks].mean())/volume.iloc[i][worst_stocks])-1
    best_vol = vol_change_best.sort_values(ascending=False)[:2].index
    worst_vol = vol_change_worst.sort_values(ascending=False)[:2].index 
    daily_update_long_returns = returns.iloc[i,:][worst_vol]
    daily_update_short_returns = 1 - (returns.iloc[i,:][best_vol]-1)
    total_daily_update_returns = pd.concat([daily_update_long_returns, daily_update_short_returns])
    returns.loc[returns.index[i], 'daily_vol'] = total_daily_update_returns.mean()

print(returns['daily_vol'].describe())
returns['daily_vol'].cumprod().plot()
print(returns.head())
# print('worst',vol_change_worst)
# print('best',vol_change_best)
```

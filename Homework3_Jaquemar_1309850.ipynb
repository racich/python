1. Please download data for the stocks of ZOOM Technologies and Zoom Video Communications from the beginning of 2019 until today (e.g. via Yahoo Finance).  Print the number of non-missing observations for each stock.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as web


# Data for Zoom Video Communications, Inc.

df_ZM = web.DataReader("ZM", 'yahoo', start='2019', end='2022')


print(df_ZM.count())

df_ZM
```

```python
# addint _ZM for identification
df_ZM_RN = df_ZM.add_suffix('_ZM')
```

```python
# Data for ZOOM Technologies, Inc.

df_ZTNO = web.DataReader("ZTNO", 'yahoo', start='2019', end='2022')

print(df_ZTNO.count())

df_ZTNO
```

```python
# adding _ZTNO for identification
df_ZTNO_RN = df_ZTNO.add_suffix('_ZTNO')
df_ZTNO_RN
```

```python
# concatenating stock data in one table
CombZZ = pd.concat([df_ZTNO_RN, df_ZM_RN], axis=1)
pd.isnull(CombZZ)
CombZZ.isnull().sum().sum()
```

```python
CombZZ
```

2. You will find that for one of the two stocks there is no data at the beginning of 2019.

a. Remove the days before the first date with non-missing data for this stock.

b. Remove the same time period from the data for the other stock so that the data for both stocks begin on the same day.

c. Rescale all prices by dividing them by the stocks' prices on the day when the time series now begin.

d. Plot the whole time series of the rescaled adjusted closing prices for both stocks in one combined graph. 

e. Compute the correlation between the two stocks' daily returns (look up the method `corr`).

```python
# 2a/b. removes days with missing values
CombZZ_rem = CombZZ.dropna()
# Removes volume column
CombZZ_rem.drop(['Volume_ZTNO', 'Volume_ZM'], axis=1, inplace=True)
CombZZ_rem
```

```python
# 2c. rescaling of all prices

CR_DIV = CombZZ_rem / CombZZ_rem.iloc[0]
CR_DIV
```

```python
# extract closing prices in another table
close_price = CR_DIV[['Adj Close_ZTNO','Adj Close_ZM']]
close_price
```

```python
# 2d. plotting of the closing prices

close_price.plot(color=['r','g'],figsize = (14,6), fontsize=12)
labels = list(close_price)
legend = plt.legend(labels, fontsize = 16)
plt.show()
```

```python
# 2d. computation of the correlation bw. the two stocks' daily returns

CombZZ["DR_ZTNO"] = CombZZ["Close_ZTNO"]/CombZZ["Open_ZTNO"]-1
CombZZ["DR_ZM"] = CombZZ["Close_ZM"]/CombZZ["Open_ZM"]-1

Daily_Returns = CombZZ.dropna()[['DR_ZTNO','DR_ZM']]
Corr = Daily_Returns.corr(method='pearson')
Corr
```

3. Consider three investment/trading strategies:

    * buy and hold a stock
    * every day buy the stock at the open sell it at the close
    * every day buy the stock at the close and sell it on the next trading day at the open

For each of the two stocks and the three strategies,

    * plot a time series of the value of a portfolio that is initially worth 100.
    * compute the total return
    * compute mean and standard deviation of daily returns

```python
#Total return for ZTNO Buy and Hold
TotalReturn_ZTNO = (CombZZ_rem['Adj Close_ZTNO'].iloc[-1] / CombZZ_rem['Open_ZTNO'].iloc[0] - 1)*100
TotalReturn_ZTNO
```

```python
#Total return for ZM Buy and Hold

TotalReturn_ZM = (CombZZ_rem['Adj Close_ZM'].iloc[-1] / CombZZ_rem['Open_ZM'].iloc[0] - 1)*100
TotalReturn_ZM
```

```python
# buy and hold strategy, assuming to buy at the open price on April 18 2019, hence setting this value at the top of the list

AdjClos = CombZZ_rem[['Adj Close_ZTNO','Adj Close_ZM']]
new_row = pd.DataFrame({'Adj Close_ZTNO':2.65, 'Adj Close_ZM':65.00}, index=['Open'])
AdjClos = pd.concat([new_row, AdjClos])
AdjClos
```

```python
# Calculation of the relative return from period n to n+1 for each stock and setting 100 as the starting value of the portfolio

AdjClos_DRT = AdjClos.div(AdjClos.shift(1)).fillna(100)

AdjClos_DRT
```

```python
# calculation of the cummulative product for the buy and hold strategy

AdjClos_DRT["AC_DRT_ZTNO"] = AdjClos_DRT["Adj Close_ZTNO"].cumprod()
AdjClos_DRT["AC_DRT_ZM"] = AdjClos_DRT["Adj Close_ZM"].cumprod()
AdjClos_DRT
```

```python
# new table for calculation and printing
AdjClos_DRT1 = AdjClos_DRT[['AC_DRT_ZTNO','AC_DRT_ZM']]
AdjClos_DRT1
```

```python
#total return BUy and Hold
AdjClos_DRT1.iloc[-1]-100
```

```python
# plot time series of the value of a portfolio 
AdjClos_DRT1.plot(color=['c','y'],figsize = (15,5), fontsize=12)
labels = list(AdjClos_DRT1)
legend = plt.legend(labels, fontsize = 16)
plt.show()
```

```python
#new table to calculate std and mean
AdjClos_DRT1 = (AdjClos.div(AdjClos.shift(1)).fillna(100)-1)[1:]
AdjClos_DRT1
```

```python
#mean deviation buy and hold 
AdjClos_DRT1.mean()
```

```python
#standard deviation buy and hold

AdjClos_DRT1.std()
```

```python
#buy open sell at close. New table with return columns

BO_SC = CombZZ_rem[['Open_ZTNO','Close_ZTNO','Open_ZM','Close_ZM']]
BO_SC["BO_SC_RZTNO"]= (BO_SC['Close_ZTNO']-BO_SC['Open_ZTNO'])/BO_SC['Open_ZTNO']+1
BO_SC["BO_SC_RZM"]= (BO_SC['Close_ZM']-BO_SC['Open_ZM'])/BO_SC['Open_ZM']+1
BO_SC
```

```python
# extraction of relevant columns

BO_SC
BO_SC_DRT1 = BO_SC[['BO_SC_RZTNO','BO_SC_RZM']]
BO_SC_DRT1
new_row1 = pd.DataFrame({'BO_SC_RZTNO':100, 'BO_SC_RZM':100}, index=['Open'])
BO_SC_DRT2 = pd.concat([new_row1, BO_SC_DRT1])
BO_SC_DRT2
```

In this exercise, you are asked to obtain some stock market data, manipulate and plot the data.

1. Please download data for the stocks of ZOOM Technologies and Zoom Video Communications from the beginning of 2019 until today (e.g. via Yahoo Finance).  Print the number of non-missing observations for each stock.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import pandas_datareader.data as web

pd.set_option('display.max_rows', 1000)

#Downloading the data
dfvideo = web.DataReader("ZM", 'yahoo', start='2019', end='2022')
dftech = web.DataReader("ZI", 'yahoo', start='2019', end='2022')
```

```python
#Non-missing observations
nm_video = len(dfvideo.index)
nm_tech = len(dftech.index)
print("The number of non-missing observations of ZM Video stock is {}.".format(nm_video))
print("The number of non-missing observations of ZI Tech stock is {}.".format(nm_tech))
```

2. You will find that for one of the two stocks there is no data at the beginning of 2019. Remove the days before the first date with non-missing data for this stock. Remove the same time period from the data for the other stock so that the data for both stocks begin on the same day.
Rescale all prices by dividing them by the stocks' prices on the day when the time series now begin. Plot the whole time series of the rescaled adjusted closing prices for both stocks in one combined graph. Compute the correlation between the two stocks' daily returns (look up the method `corr`).

```python
#Data of both stocks beginning on the same day
dfvideo=dfvideo.drop(dfvideo.index[:(len(dfvideo.index)-len(dftech.index))])

#Rescaling video stock
dfvideo_rescaled=dfvideo/dfvideo.iloc[0]
dfvideo_rescaled
```

```python
#Rescaling tech stock
dftech_rescaled=dftech/dftech.iloc[0]
dftech_rescaled
```

```python
#Combining the rescaled adj. closing price of both stocks in one data frame

dfvideo_rescaled=dfvideo_rescaled.rename(columns={'Adj Close': 'ZM'})
dftech_rescaled=dftech_rescaled.rename(columns={'Adj Close': 'ZI'})

video = dfvideo_rescaled['ZM']
tech = dftech_rescaled['ZI']

result = pd.concat([video, tech], axis = 1)
result
```

```python
#Plot of the time series of rescaled adj. closing prices
result.plot(figsize=(12,6))

plt.title('Rescaled adjusted closing prices for ZM and ZI', fontsize=20)
plt.show()
```

```python
#Find daily returns
returns = (result / result.shift(1)).iloc[1:]

#Correlation
correlation = returns.corr().iloc[0]['ZI']
print("The correlation between the stocks is {}.".format(correlation))
```

3. Consider three investment/trading strategies:

* buy and hold a stock
* every day buy the stock at the open sell it at the close
* every day buy the stock at the close and sell it on the next trading day at the open

For each of the two stocks and the three strategies,

* plot a time series of the value of a portfolio that is initially worth 100.
* compute the total return
* compute mean and standard deviation of daily returns

```python
# Buy-Hold Strategy

START_CAPITAL = 100

#Time evolution
stock_ts = pd.concat([dfvideo_rescaled['Close'], dftech_rescaled['Close']], axis = 1)
stock_ts.columns = ['ZM', 'ZI']
stock_ts *= 100

stock_ts.plot()

#Daily returns

stock_ts_returns = (stock_ts / stock_ts.shift(1)).iloc[1:]
print(f"Mean daily returns: \n{stock_ts_returns.mean(axis = 0)}\n")
print(f"Standard deviation of daily returns: \n{stock_ts_returns.std(axis = 0)}\n")

# Total return in percent
buying_price_tech = dftech.iloc[0]['Open']
selling_price_tech = dftech.iloc[-1]['Close']
buying_price_video = dfvideo.iloc[0]['Open']
selling_price_video = dfvideo.iloc[-1]['Close']
total_return_tech =  selling_price_tech / buying_price_tech
total_return_video = selling_price_video / buying_price_video 

print(f"Total Returns\nZI: {total_return_tech} ZM: {total_return_video} ")
```

```python
#Buy at Open / Sell at Close

#Divide Close by Open price for each day
daily_returns = pd.concat([
    dfvideo['Close'] / dfvideo['Open'],
    dftech['Close'] / dftech['Open']],
    axis = 1
)

print(f"Mean daily returns: \n{daily_returns.mean(axis = 0)}\n")
print(f"Standard deviation of daily returns: \n{daily_returns.std(axis = 0)}\n")


daily_returns.columns = ['ZM', 'ZI']
stock_ts_open_close = daily_returns.cumprod()*100
stock_ts_open_close.plot()

#Total return 
total_returns = stock_ts_open_close.iloc[-1] / 100
print(f"Total Returns:\nZI {total_returns['ZI']} ZM: {total_returns['ZM']} ")
```

```python
# Buy at close sell at open

#Divide Open by close price from the day before
daily_returns_close_open = pd.concat([
    (dfvideo['Open'].shift(-1) / dfvideo['Close']).iloc[:-1],
    (dftech['Open'].shift(-1) / dftech['Close']).iloc[:-1]],
    axis = 1
)

daily_returns_close_open.columns = ['ZM', 'ZI']

print(f"Mean daily returns: \n{daily_returns_close_open.mean(axis = 0)}\n")
print(f"Standard deviation of daily returns: \n{daily_returns_close_open.std(axis = 0)}\n")

stock_ts_close_open = daily_returns_close_open.cumprod()*100
stock_ts_close_open.plot()

#Total return 
total_returns_close_open = stock_ts_close_open.iloc[-1] / 100
print(f"Total Returns: \nZI {total_returns_close_open['ZI']} ZM: {total_returns_close_open['ZM']} ")
```
